Q1.Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import plotly.graph_objects as go
```

```python
import seaborn as sns
import plotly.express as px

titanic_data = sns.load_dataset("titanic")

fig = px.scatter(titanic_data, x='age', y='fare', color='class', hover_data=['sex', 'survived'],
                 title='Scatter Plot for Age and Fare in Titanic Dataset',
                 labels={'age': 'Age', 'fare': 'Fare', 'class': 'Passenger Class'},
                 size_max=10)

fig.show()
```

Q 2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import seaborn as sns
import plotly.express as px

tips_data = sns.load_dataset("tips")

# Create a box plot using Plotly Express
fig = px.box(tips_data, x='day', y='total_bill', color='sex', notched=True,
             title='Box Plot for Total Bill by Day and Sex',
             labels={'day': 'Day of Week', 'total_bill': 'Total Bill', 'sex': 'Sex'})

fig.show()
```

Q 3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python

tips_data = sns.load_dataset("tips")

fig = px.histogram(tips_data, x='sex', y='total_bill', color='day', barmode='group',
                   pattern_shape='smoker',
                   title='Histogram for Total Bill by Sex with Day and Smoker',
                   labels={'sex': 'Sex', 'total_bill': 'Total Bill', 'day': 'Day of Week', 'smoker': 'Smoker'},
                   category_orders={'day': ['Thur', 'Fri', 'Sat', 'Sun']})

fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python

iris_data = sns.load_dataset("iris")

fig = px.scatter_matrix(iris_data, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                        color='species', symbol='species',
                        title='Scatter Matrix Plot for Iris Dataset',
                        labels={'sepal_length': 'Sepal Length', 'sepal_width': 'Sepal Width',
                                'petal_length': 'Petal Length', 'petal_width': 'Petal Width'})

fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset from Seaborn
tips_data = sns.load_dataset("tips")

# Create a distplot using Seaborn
sns.distplot(tips_data['total_bill'], kde=True, hist=True)

# Add title and labels to the plot
plt.title("Distribution of Total Bill Amount")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import plotly.express as px

# Load the "tips" dataset from Seaborn
tips_data = sns.load_dataset("tips")

# Create a histogram using Plotly Express
fig = px.histogram(tips_data, x='total_bill', nbins=30, marginal='rug',
                   title='Distribution of Total Bill Amount',
                   labels={'total_bill': 'Total Bill', 'count': 'Frequency'})

# Show the plot
fig.show()
```
